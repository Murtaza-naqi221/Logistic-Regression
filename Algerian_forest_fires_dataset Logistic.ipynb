```python
from numpy import mean
from numpy import std
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
df = pd.read_excel('Algerian_forest_fires_dataset_UPDATE.xlsx')
df.head(10)
```

```python
len(df)
```

```python
X = df.drop('Classes',axis=1)
y = df['Classes']
```

```python
X = X.drop('day',axis=1)
X = X.drop('month',axis=1)
X = X.drop('year',axis=1)
X
```

```python
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
y == label_encoder.fit_transform(y)
```

```python
pd.to_numeric(X['Temperature'], errors='coerce')
X
```

```python
from sklearn import preprocessing
d = preprocessing.normalize(X, axis=0)
X = pd.DataFrame(d)
X.head()
```

```python
X, y = make_classification(n_samples=240, n_features=7, n_informative=2, n_redundant=1, random_state=1)
# prepare the cross-validation procedure
cv = KFold(n_splits=10, random_state=1, shuffle=True)
# create model
model = LogisticRegression(penalty='l2', dual=False, tol=0.01, C=1.0)
# evaluate model
scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```
